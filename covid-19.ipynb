**<h2>Total States Data</h2>**

```python
# loopng through the data for the total states data
states_total = []

for i in states.keys():
    if i not in 'TT':
        state = i
        population = int(states[i]['meta']['population'])
        deceased = int(states[i]['total']['deceased'])
        recovered = int(states[i]['total']['recovered'])
        tested = int(states[i]['total']['tested'])
        vaccinated1 = int(states[i]['total']['vaccinated1'])
        vaccinated2 = int(states[i]['total']['vaccinated2'])

        dictionary_state_total = {'state' : dic_state_name[state], 'state_short' : state, 'state_total_population' : population, 'state_total_deceased' : deceased, 'state_total_recovered' : recovered, 'state_total_tested' : tested, 'state_total_vaccinated_1' : vaccinated1, ' state_total_vaccinated_2' :vaccinated2}
        states_total.append(dictionary_state_total)
```

```python
# creating dataframe of states total

states_total_df = pd.DataFrame(states_total)
print(len(states_total_df))
states_total_df.head(5)
```

```python
# converting to csv states total
states_total_df.to_csv('states_total_df.csv',index=False)
```

**<h2>States Delta</h2>**

```python
#  loopng through the data for delta of states

states_delta = []
for i in states.keys():
    if i != 'TT':
        state = i
        confirmed = 0
        recovered = 0
        deceased = 0
        tested = 0
        vaccinated1 = 0
        vaccinated2 = 0
        if 'confirmed' in states[i]['delta']:
            confirmed += int(states[i]['delta']['confirmed'])

        if 'recovered' in states[i]['delta']:
            recovered += int(states[i]['delta']['recovered'])

        if 'deceased' in states[i]['delta']:
            deceased += int(states[i]['delta']['deceased'])

        if 'tested' in states[i]['delta']:
            tested += int(states[i]['delta']['tested'])

        if 'vaccinated1' in states[i]['delta']:
            vaccinated1 += int(states[i]['delta']['vaccinated1'])

        if 'vaccinated2' in states[i]['delta']:
            vaccinated2 += int(states[i]['delta']['vaccinated2'])

        dictionary_state_delta = {'state' : dic_state_name[state], 'state_short' : state, 'state_delta_confirmed' : confirmed, 'state_delta_recovered' : recovered, 'state_delta_deceased' : deceased, 'state_delta_tested' : tested, 'state_delta_vaccinated_1' : vaccinated1, 'state_delta_vaccinated_2' : vaccinated2}
        states_delta.append(dictionary_state_delta)
```

```python
# creating dataframe of states delta
states_delta_df = pd.DataFrame(states_delta)
print(len(states_delta_df))
states_delta_df.head()
```

```python
# converting to csv states delta
states_delta_df.to_csv('states_delta_df.csv',index=False)
```

**<h2>TimeSeries Link Data</h2>**

```python
#Assigning the link
timeseries_data = requests.get('https://data.covid19india.org/v4/min/timeseries.min.json')
```

```python
# Creating json
timeseries_data
```

```python
# Creating json
timeseries = timeseries_data.json()
```

##  total of date

```python
print('Total number of states in data : ',len(timeseries))
```

```python
#  name of states in timeseries
timeseries.keys()
```

```python
# looping through data to create table like form for TOTAL key of each states date


date_total = []
for i in timeseries.keys():
    if i not in ('TT','UN'):
        for j in timeseries[i]:
            for k in timeseries[i][j]:
                state_time = i
                date = k
                confirmed = 0
                recovered = 0
                deceased = 0
                tested = 0
                vaccinated1 = 0
                vaccinated2 = 0
                if 'total' in timeseries[i][j][k] and i not in ('TT','UN'):
                    if 'confirmed' in timeseries[i][j][k]['total']:
                        confirmed += int(timeseries[i][j][k]['total']['confirmed'])

                    if 'recovered' in timeseries[i][j][k]['total']:
                        recovered += int(timeseries[i][j][k]['total']['recovered'])

                    if 'deceased' in timeseries[i][j][k]['total']:
                        deceased += int(timeseries[i][j][k]['total']['deceased'])

                    if 'tested' in timeseries[i][j][k]['total']:
                        tested += int(timeseries[i][j][k]['total']['tested'])

                    if 'vaccinated1' in timeseries[i][j][k]['total']:
                        vaccinated1 += int(timeseries[i][j][k]['total']['vaccinated1'])

                    if 'vaccinated2' in timeseries[i][j][k]['total']:
                        vaccinated2 += int(timeseries[i][j][k]['total']['vaccinated2'])

                    dictionary_total = {'state' : dic_state_name[state_time], 'state_short' : state_time, 'date' : date, 'total_confirmed' : confirmed, 'total_recovered' : recovered, 'total_deceased' : deceased, 'total_tested' : tested, 'total_vaccinated_1' : vaccinated1, 'total_vaccinated_2' : vaccinated2}

                    date_total.append(dictionary_total)
```

```python
# creating dataframe of date total and viewing
date_total_df = pd.DataFrame(date_total)
print(len(date_total_df))
date_total_df.tail()
date_total_df.head()
```

```python
# Converting date total to csv
date_total_df.to_csv('date_total_df.csv',index = False)
```
